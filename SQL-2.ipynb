## a.  One to many relationship
1. Report the account representative for each customer.
2. Report total payments for Atelier graphique.
3. Report the total payments by date
4. Report the products that have not been sold.
5. List the amount paid by each customer.
6. How many orders have been placed by Herkku Gifts?
7. Who are the employees in Boston?
8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.
9. List the value of 'On Hold' orders.
10.Report the number of orders 'On Hold' for each customer.

```python
import pymysql
import pandas as pd
host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host = host,
    port = int(port),
    user = user,
    passwd = password,
    db = database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# 1. Report the account representative for each customer.
run("""
SELECT
    customerName,
    CONCAT(firstName," ",lastName) AS repsEmployee
FROM
    customers t1
    INNER JOIN
    employees t2
        ON t1.salesRepEmployeeNumber = t2.employeeNumber
ORDER BY
    customerName
""")
```

```python
# 2. Report total payments for Atelier graphique.
run("""
SELECT
    customerName,
    SUM(amount) AS totalPayments
FROM
    customers t1
    INNER JOIN
    payments t2
        ON t1.customerNumber = t2.customerNumber
WHERE
    customerName = 'Atelier graphique'
""")
```

```python
# 3. Report the total payments by date
run("""
SELECT
    paymentDate,
    SUM(amount) AS totalPayments
FROM
    payments
GROUP BY
    paymentDate
""")
```

```python
# 4. Report the products that have not been sold.
run("""
SELECT
    productName,
    quantityOrdered
FROM
    orderdetails t1
    RIGHT JOIN
    products t2
        ON t1.productCode = t2.productCode
WHERE
    quantityOrdered is Null
""")
```

```python
# 5. List the amount paid by each customer.
run("""
SELECT
    customerName,
    SUM(amount) AS amountPaid
FROM
    customers t1
    INNER JOIN
    payments t2
        ON t1.customerNumber = t2.customerNumber
GROUP BY
    customerName
""")
```

```python
# 6. How many orders have been placed by Herkku Gifts?
run("""
SELECT
    customerName,
    SUM(orderNumber) AS totalOrder
FROM
    customers t1
    INNER JOIN
    orders t2
        ON t1.customerNumber = t2.customerNumber
WHERE
    customerName = 'Herkku Gifts'
""")
```

```python
# 7. Who are the employees in Boston?
run("""
SELECT
    CONCAT(firstName," ",lastName) AS employeeName,
    city
FROM
    employees t1
    INNER JOIN
    offices t2
        ON t1.officeCode = t2.officeCode
WHERE
    city = 'Boston'
""")
```

```python
#8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.
run("""
SELECT
    customerName,
    amount
FROM
    customers t1
    INNER JOIN
    payments t2
        ON t1.customerNumber = t2.customerNumber
WHERE
    amount>100000
ORDER BY
    amount DESC
""")
```

```python
# 9. List the value of 'On Hold' orders.
run("""
SELECT
    customerNumber,
    (quantityOrdered*priceEach) AS Value
FROM 
    orders t1
    INNER JOIN
    orderdetails t2
        ON t1.orderNumber = t2.orderNumber 
WHERE 
    status='On Hold'
""")
```

```python
# 10.Report the number of orders 'On Hold' for each customer.
run("""
SELECT
    customerName,
    COUNT(status) AS On_Hold
FROM
    customers t1
    INNER JOIN
    orders t2
        ON t1.customerNumber = t2.customerNumber
WHERE
    status='On Hold'
GROUP BY
    customerName
""")
```

## b. Many to many relationship
1. List products sold by order date.
2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
4. Are there any products that appear on all orders?
5. List the names of products sold at less than 80% of the MSRP.
6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
7. List the products ordered on a Monday.
8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
# 1. List products sold by order date.
# run("""
# SELECT
#     productName,
#     orderDate,
#     status
# FROM
#     ((orderdetails
#         INNER JOIN
#     orders USING (orderNumber))
#         INNER JOIN
#     products USING (productCode))
# WHERE
#     status = 'Shipped '
# ORDER BY
#     orderDate
# """)

#atau

run("""
SELECT 
    productName,    
    orderDate,
    status
FROM
    ((orderdetails t1
    INNER JOIN
    orders t2 
        ON t1.orderNumber = t2.orderNumber)
    INNER JOIN
    products t3 
        ON t1.productCode = t3.productCode)  
WHERE
    status = 'Shipped'
ORDER BY 
    orderDate 
""")
```

```python
# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
# run("""
# SELECT
#     productName,
#     orderDate
# FROM
#     ((orderdetails t1
#     INNER JOIN
#     orders t2
#         ON t1.orderNumber = t2.orderNumber)
#     INNER JOIN
#     products t3
#         ON t1.productCode =t3.productCode)
# WHERE
#     productName = '1940 Ford Pickup Truck'
# ORDER BY
#     orderDate DESC
# """)

#atau

run("""
SELECT
    productName,
    orderDate
FROM
    ((orderdetails
        INNER JOIN
    orders USING (orderNumber))
        INNER JOIN
    products USING (productCode))
WHERE
    productName = '1940 Ford Pickup Truck'
ORDER BY
    orderDate DESC
""")
```

```python
# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
run("""
SELECT
    customerName,
    orderNumber,
    amount
FROM
    ((customers
        INNER JOIN
    orders USING(customerNumber))
        INNER JOIN
    payments USING(customerNumber))
WHERE
    amount > 25000
ORDER BY
    orderNumber
""")
```

```python
# 4. Are there any products that appear on all orders?
run("""
SELECT 
    productName,
    COUNT(productCode) AS productAllAppear
FROM 
    orderdetails 
        INNER JOIN
    products USING(productCode)
GROUP BY 
    productCode 
HAVING 
    productAllAppear = (SELECT COUNT(*) FROM orders)
""")
```

```python
# 5. List the names of products sold at less than 80% of the MSRP.
run("""
SELECT
    productName,
    MSRP,
    orderNumber,
    status,
    priceEach,
    (priceEach / MSRP * 100) AS sellPricePercentageOfTheMSRP
FROM
    ((orderdetails
        INNER JOIN
    products USING (productCode))
        INNER JOIN
    orders USING (orderNumber))
WHERE
    priceEach < 80 / 100 * MSRP
        AND
    status = 'Shipped'
""")
```

```python
# 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
run("""
SELECT
    productName,
    buyPrice,
    orderNumber,
    status,
    priceEach,
    ((priceEach - buyPrice) / buyPrice * 100) AS markupPercentage
FROM
    ((orderdetails
        INNER JOIN
    products USING (productCode))
        INNER JOIN
    orders USING (orderNumber))
WHERE
    priceEach >= 2 * buyPrice
        AND
    status = 'Shipped'
""")
```

```python
# 7. List the products ordered on a Monday.
run("""
SELECT 
    productName,
    DAYNAME(orderDate) AS Hari,
    orderDate
FROM
    ((orderdetails t1
    INNER JOIN
    products t2 
        ON t1.productCode = t2.productCode)
    INNER JOIN
    orders t3 
        ON t1.orderNumber = t3.orderNumber)  
WHERE 
    DAYNAME(orderDate)='Monday'
""")
```

```python
# 8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
SELECT 
    productName,
    status
FROM
    ((orderdetails t1
    INNER JOIN
    products t2 
        ON t1.productCode = t2.productCode)
    INNER JOIN
    orders t3 
        ON t1.orderNumber = t3.orderNumber)  
WHERE 
    status='On Hold'
""")
```
