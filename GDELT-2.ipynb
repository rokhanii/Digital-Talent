```python
data2 = data.drop(data.loc[:,'Jumlah Penduduk Miskin Menurut Provinsi (Ribu Jiwa)':'Unnamed: 43'].head(0).columns, axis=1)
data2.head()
```

```python
data3 = data2.drop(columns=['Unnamed: 54'])
data3
```

```python
data4 = data3.dropna()
data4.tail()
```

```python
data5 = data4.iloc[:-1]
data5.tail()
```

```python
data5.columns = ['Provinsi','2013/1','2013/2','2014/1','2014/2', '2015/1','2015/2', '2016/1','2016/2','2017/1','2017/2']
data5.tail()
```

```python
ave_data = data5.copy()
ave_data['2013/1'] = pd.to_numeric(ave_data['2013/1'])
ave_data['2013/2'] = pd.to_numeric(ave_data['2013/2'])
ave_data['2014/1'] = pd.to_numeric(ave_data['2014/1'])
ave_data['2014/2'] = pd.to_numeric(ave_data['2014/2'])
ave_data['2015/1'] = pd.to_numeric(ave_data['2015/1'])
ave_data['2015/2'] = pd.to_numeric(ave_data['2015/2'])
ave_data['2016/1'] = pd.to_numeric(ave_data['2016/1'])
ave_data['2016/2'] = pd.to_numeric(ave_data['2016/2'])
ave_data['2017/1'] = pd.to_numeric(ave_data['2017/1'])
ave_data['2017/2'] = pd.to_numeric(ave_data['2017/2'])
ave_data['ave2013'] = ave_data[['2013/1', '2013/2']].mean(numeric_only = True, axis = 1)
ave_data['ave2014'] = ave_data[['2014/1', '2014/2']].mean(numeric_only = True, axis = 1)
ave_data['ave2015'] = ave_data[['2015/1', '2015/2']].mean(numeric_only = True, axis = 1)
ave_data['ave2016'] = ave_data[['2016/1', '2016/2']].mean(numeric_only = True, axis = 1)
ave_data['ave2017'] = ave_data[['2017/1', '2017/2']].mean(numeric_only = True, axis = 1)
ave_data
```

```python
data6 = ave_data.drop(ave_data.loc[:,'2013/1':'2017/2'].head(0).columns, axis=1)
data6
```

```python
data6.columns = ['Provinsi','2013','2014','2015','2016','2017']
data6.tail()
```

```python
data_kemiskinan = data6.set_index('Provinsi')
data_kemiskinan
```

```python
# import packages
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

fig1 = plt.figure()

%matplotlib notebook

# load data
# revels_data = pd.read_csv("directory.data.txt")
# rd = revels_data
data_kemiskinanfix = pd.DataFrame(data_kemiskinan,columns=['2013', '2014', '2015', '2016', '2017'])
ax = data_kemiskinanfix.plot(stacked=True)

# grouped bar plot
grouped = ax.groupby(['Provinsi', '2013', '2014', '2015', '2016', '2017'])["Contents"].sum().unstack().fillna(0)
grouped.plot(kind="bar", stacked=True, color=["#a2653e", "#a6814c", "#fd5956", "#fd8d49", "#9c6da5", "#c9ae74"])

# title and axis labels
plt.title("NUMBER OF REVELS PER PACKET", weight="bold")
plt.ylabel("Contents")
plt.xlabel("Packet")

# plot median line
median = 19
plt.axhline(median, color='grey', linestyle='dashed', linewidth=0.5)

# legend properties
plt.legend(loc=2, prop={"size":7})

# extend y axis to fit legend in
axes = plt.gca()
axes.set_ylim([0,30])

# show plot
plt.show()
```
