```python
import os
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
# 1. Prepare a list of offices sorted by country, state, city.
run("SELECT * FROM `offices` ORDER BY country,state,city")
```

```python
#2. How many employees are there in the company?
run("SELECT COUNT(employeeNumber) as JumlahPegawai FROM `employees`")
```

```python
#3. What is the total of payments received?
run("SELECT SUM(amount) as JumlahPembayaran FROM `payments`")
```

```python
#4. List the product lines that contain 'Cars'.
run("SELECT * FROM `productLines` WHERE productLine LIKE '%Cars'")
```

```python
#5. Report total payments for October 28, 2004.
run("SELECT SUM(amount) AS JumlahPembayaran FROM `payments` WHERE paymentDate <= '2004-10-28' ")
```

```python
#6. Report those payments greater than $100,000.
run("SELECT * FROM `payments` WHERE amount > 100000")
```

```python
#7. List the products in each product line.
run("SELECT productLine, productName FROM products ORDER BY productLine ")
```

```python
#8. How many products in each product line?
run("SELECT productLine, COUNT(productName) AS JumlahProduk FROM `products` GROUP BY productLine ")
```

```python
#9. What is the minimum payment received?
run("SELECT MIN(amount) AS HargaTermurah FROM `payments`")
```

```python
#10. List all payments greater than twice the average payment.
run("SELECT * FROM payments WHERE amount >= (SELECT 2*AVG(amount) FROM payments)")
```

```python
#11. What is the average percentage markup of the MSRP on buyPrice?
run("SELECT AVG((MSRP-buyPrice)*100/buyPrice) as RerataMarkUp FROM `products`")
```

```python
#12. How many distinct products does ClassicModels sell?
run ("SELECT COUNT(Distinct(productName)) as DistinctClassicModel FROM products WHERE productLine Like '%Classic%' ")
```

```python
#13. Report the name and city of customers who don't have sales representatives?
run("""
SELECT customerName AS NamaKonsumen, city AS Kota
FROM customers
WHERE salesRepEmployeeNumber is null
""")
```

```python
#14. What are the names of executives with VP or Manager in their title?Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.
run("""
    SELECT
    CONCAT(firstName,' ',lastName) as Nama, jobTitle
    FROM employees
    WHERE jobTitle like '%VP%' OR jobTitle like '%Manager%'
    """)
```

```python
#15. Which orders have a value greater than $5,000
# run("SELECT * FROM `payments` WHERE amount > 100000")
run("""
SELECT orderNumber, 
productCode,
(quantityOrdered*priceEach) AS Value
FROM orderdetails
WHERE (quantityOrdered*priceEach) > 5000 """)
```
